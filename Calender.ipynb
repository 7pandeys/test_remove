```python
from datetime import datetime as dt
```

```python
t = dt.today()
```

```python
import pandas as pd

Year,Age,Task = [],[],[]
for i in range(30):
    Year.append(t.year+i)
    Age.append(26+i)
    Task.append(input(f'Year:{Year[i]}, Age:{Age[i]}, Task:{i} '))

df = pd.DataFrame()
df['Year'] = Year
df['Age'] = Age
df['Task'] = Task 
```
